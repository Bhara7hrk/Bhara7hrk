### 3.0  Display the percentage distribution of the data in each year using the pie chart.

```python
size = df['year'].value_counts()/len(df)*100
plt.pie(size, labels= size.index, autopct='%1.1f%%', startangle=0,shadow = True,explode =(0.2,0.2,0.2,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05))
plt.axis('equal')
plt.show()
```

- Distribution of data in each year can be seen in above chart.
- distribution among each year is almost equal (6.4% to 7%)
- 2003,2002,2001 is highest with 7% and 2009,2010,2011,2012 is lowest with 6.4%

### 5.0 Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

### 6.0   Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x=df["year"], y=df["prodvalue"], data=df,palette="Set3")
plt.show()
```

- The median value is almost similar in each year.
- There are several outliers present in all the years.
- The hight prodvalue among all the years is 2010.
***

### 7.0 Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
sns.pairplot(df, hue='year', vars=['numcol','yieldpercol','totalprod','prodvalue'],palette='Set1')
plt.show()
```

- Strong positive relation can be seen between prodvalue and totalprod.
- Strong positive relation can be seen between prodvalue and numcol.
- Strong positive relation can be seen between numcol and Totalprod.
- The distribution plot shows variation of the data in each year.
- There is no relation between yieldpercol and productval and numcol.

### 8.0 Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
cor = df[['numcol', 'yieldpercol','totalprod','stocks','priceperlb','prodvalue']].corr()
cor
```

```python
sns.heatmap(cor,annot=True,cmap="YlGnBu")
```

- From the above heat map we can see the co eff of correlation between the variables.
- we can see strong positive relation between totalprod, prodval, numcol, stocks.
- there is a weak/ no relation between priceperlb and  yieldpercol with other variables.
